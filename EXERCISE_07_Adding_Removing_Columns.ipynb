```python
import pandas as pd
import numpy as np
import os
```

```python
os.listdir('../data')
```

```python
#Import the JoeBidenTweets.csv dataset.  Use the tweet "id" as the index

tweets = pd.read_csv('../data/JoeBidenTweets.csv', index_col='id')
tweets.head()
```

```python
#Drop the "url" column (NOT inplace)

tweets.drop(columns='url')
```

```python
361388562 in tweets.index
```

```python
tweets.loc[361388562]
```

```python
#Drop the row with the id of 361388562(NOT inplace)

tweets.drop(labels=361388562, axis=0).head()
```

```python
#Add a new column called "user".  Set it to "Joe Biden" for all rows

tweets.insert(1, 'user', 'Joe Biden')
tweets.head()
```

```python
#Create a new column called "ratio".  It should contain the result of dividing each tweet's replies by its retweets

tweets.insert(6, 'ratio', tweets['replies'] / tweets['retweets'])
tweets.head()
```

```python
#Find the top 10 highest ratio tweets

tweets.sort_values('ratio', ascending=False).head(10)
```

```python
# An alternate way of getting top 10 highest ratio tweets

tweets['ratio'].nlargest(10)
```

```python
# Another way of getting top 10 highest ratio tweets

tweets.sort_values('ratio', ascending=False)['ratio'].nlargest(10)
```

```python
#Create a column called "interactions" that contains the sum of each tweet's replies, retweets, quotes, and likes

tweets['interactions'] = tweets['replies'] + tweets['retweets'] + tweets['quotes'] + tweets['likes']
tweets.head()
```

```python
# Round the ratio of tweets to 2 decimal places

tweets['ratio']= round(tweets['ratio'], 2)
tweets.head()
```

```python
#Find the top 8 tweets by number of "interactions"

tweets.sort_values('interactions', ascending=False).head(8)
```
