```python
import pandas as pd
import numpy as np
import os
```

```python
bestsellers = pd.read_csv('../data/bestsellers.csv')
bestsellers.head()
```

```python
bestsellers.info()
```

```python
#Find the books written by Pete Souza

bestsellers[bestsellers['Author'] == "Pete Souza"]
```

```python
# Find the books that are under 10 dollars

bestsellers[bestsellers['Price'] < 10]
```

```python
# Get the count of the books that are under 10 dollars

bestsellers[bestsellers['Price'] < 10]['Price'].count()
```

```python
#Find the books that have a price between 50 and 60 dollars

bestsellers[bestsellers['Price'].between(50, 60)]
```

```python
#Find all the books written by Kristin Hannah, Andy Weir, or Delia Owens

authors = ['Kristin Hannah', 'Andy Weir', 'Delia Owens']

bestsellers[bestsellers['Author'].isin(authors)]

# bestsellers[]
```

```python
#Find the Non Fiction books that are rated 4.9

fiction = bestsellers['Genre'] == 'Non Fiction'
rating = bestsellers['User Rating'] == 4.9

bestsellers[fiction & rating]
```

```python
#Find the number of Non Fiction books that are rated 4.9

bestsellers[fiction & rating]['Genre'].value_counts()
```

```python
#Find the fiction book with the lowest User Rating

bestsellers['User Rating'].nsmallest(1)
bestsellers.iloc[353]
```

```python
#Find 2012's top 5 Fiction books with the most Reviews

year_2021 = bestsellers['Year'] == 2012
fiction = bestsellers['Genre'] == 'Fiction'

year_2012_fictions = bestsellers[year_2021 & fiction]
year_2012_fictions.sort_values('Reviews', ascending=False).head()
```

```python
# Create a bar plot showing the 5 authors who have the most books with a rating under 4.5

ratings_below = bestsellers[bestsellers['User Rating'] < 4.5]
ratings_below.sort_values('Author', ascending=False)['Author'].value_counts().head().plot(kind='bar')
```
